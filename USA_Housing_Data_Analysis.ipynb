```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('usa_housing_kaggle.csv')

# Display basic information
print("Dataset Info:")
print(df.info())
print("\
Basic Statistics:")
print(df.describe())
print("\
Missing Values:")
print(df.isnull().sum())
```

Based on the analysis of the housing dataset, here are the key findings:

Data Quality:

The dataset is clean with no missing values
Contains 300 houses with 10 features including price, physical characteristics, and location metrics
Key Statistics:
Average Price: $522,761.92

Median Price: $505,806.50

Price Range: $112,666.00 - $999,684.00

Average House Characteristics:
Average Square Feet: 3,052.75

Average Bedrooms: 2.97

Average Bathrooms: 2.46

## Visualizations:

## 1. Correlation Heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

Strong positive correlation between price and square feet.
Moderate positive correlation between price and school rating.
Negative correlation between price and crime rate.
## 2. Price Distribution:

```python
# Create distribution plots for Price
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Price', bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.show()
```

The price distribution shows a slight right skew.
Most houses are clustered in the $300,000-$700,000 range.
## Key Relationships:

```python
# Create scatter plots for key relationships
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

sns.scatterplot(data=df, x='SquareFeet', y='Price', ax=axes[0,0])
axes[0,0].set_title('Price vs Square Feet')

sns.scatterplot(data=df, x='YearBuilt', y='Price', ax=axes[0,1])
axes[0,1].set_title('Price vs Year Built')

sns.scatterplot(data=df, x='CrimeRate', y='Price', ax=axes[1,0])
axes[1,0].set_title('Price vs Crime Rate')

sns.scatterplot(data=df, x='SchoolRating', y='Price', ax=axes[1,1])
axes[1,1].set_title('Price vs School Rating')

plt.tight_layout()
plt.show()

```

Clear positive relationship between square footage and price.
Newer homes tend to have higher prices.
Higher crime rates generally associated with lower prices.
Better school ratings tend to correlate with higher prices.
